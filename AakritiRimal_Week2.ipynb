```python
# Adding Some Missing Values
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris() # Load the Iris dataset
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
np.random.seed(42) # Introduce missing values randomly
mask = np.random.rand(*iris_df.shape) < 0.1 # 10%
iris_df[mask] = np.nan
print("Missing Values in Iris Dataset:")
print(iris_df.isnull().sum())

# Filling missing values with forward fill (ffill), mean, median, and 0
iris_df_ffill = iris_df.ffill()
iris_df_mean = iris_df.fillna(iris_df.mean())
iris_df_median = iris_df.fillna(iris_df.median())
iris_df_zero = iris_df.fillna(0)
# Expand iris_df with filled columns
iris_df_expanded = pd.concat([iris_df, iris_df_ffill.add_suffix('_ffill'), iris_df_mean.add_suffix('_mean'),iris_df_median.add_suffix('_median'),iris_df_zero.add_suffix('_zero')], axis=1)
# Display the head of the expanded DataFrame
print("\nDataset after Filling Missing Values:")
print(iris_df_expanded.head())
```

```python
#------------------Pivoting----------------------------------------
import pandas as pd
# Sample DataFrame
data = {'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
'City': ['Kathmandu', 'Pokhara', 'Kathmandu', 'Pokhara'],
'Temperature': [15, 18, 16, 19]}
df = pd.DataFrame(data)
# Pivot: Reshape data to show cities as columns
pivoted_df = df.pivot(index='Date', columns='City', values='Temperature')
print(pivoted_df)

# Melt: Convert wide data back to long format
melted_df = pd.melt(pivoted_df.reset_index(), id_vars=['Date'],
var_name='City', value_name='Temperature')

print(melted_df)
```

```python
import pandas as pd
# Sample DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
# Row-wise concatenation
combined_rows = pd.concat([df1, df2], axis=0)
print("Row-wise concatenation:")
print(combined_rows)
# Column-wise concatenation
combined_cols = pd.concat([df1, df2], axis=1)
print("\nColumn-wise concatenation:")
print(combined_cols)

# Sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Score': [85, 90, 88]})
# Inner join
inner_merged = pd.merge(df1, df2, on='ID', how='inner')
print("Inner Join:")
print(inner_merged)
# Left join
left_merged = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(left_merged)
# Outer join
outer_merged = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:")
print(outer_merged)
```

```python
import pandas as pd

#1
file_path = "/content/drive/MyDrive/Datasets/medical_students_dataset.csv"
medical_df = pd.read_csv(file_path)

#2
print("Dataframe info: ")
print(medical_df.info())

print("\nColumns with missing values: ")
missing_values = medical_df.isnull().sum()
print(missing_values[missing_values > 0])

#3
#Filling numerical data with mean because it is suitable for numerical data as it maintains overall distribution
if 'numerical_column' in medical_df.columns:
  medical_df['numerical_column'].fillna(medical_df['numerical_column'].mean(), inplace=True)
#Filling categorical data with mode as it is appropriate for it as it preserves common trends
if 'categorical_column' in medical_df.columns:
  medical_df['categorical_column'].fillna(medical_df['categorical_column'].mode()[0], inplace=True)

print("\nAfter filling missing values: ")
print(medical_df.isnull().sum())

#4
duplicates_count = medical_df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates_count}")
if duplicates_count > 0:
  medical_df.drop_duplicates()
  print("\nDuplicate rows removed.")
print("\nFirst few rows of the cleaned dataset:")
print(medical_df.head())
```

3.2 Exercises - Data Cleaning and Transformations with "Titanic Datasets:"

```python
import pandas as pd
import matplotlib.pyplot as plt

#Problem-1
file_path = "/content/drive/MyDrive/Datasets/Titanic-Dataset.csv"
titanic_df = pd.read_csv(file_path)
#Creating subset dataframe
columns_subset = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
subset_df = titanic_df[columns_subset]
#Retain row where pclass is 1
first_class_df = subset_df[subset_df['Pclass'] == 1]
fare_mean = first_class_df['Fare'].mean()
fare_median = first_class_df['Fare'].median()
fare_max = first_class_df['Fare'].max()
fare_min = first_class_df['Fare'].min()
print("Fare Statistics for First-Class Passengers:")
print(f"Mean: {fare_mean}")
print(f"Median: {fare_median}")
print(f"Maximum: {fare_max}")
print(f"Minimum: {fare_min}")

#Problem-2
#check null values for age column
null_age_count = first_class_df['Age'].isnull().sum()
print(f"Number of null values in 'Age' column: {null_age_count}")
#drop null value
first_class_cleaned = first_class_df.dropna(subset=['Age'])
print("\nNull values in 'Age' column after dropping rows:")
print(first_class_cleaned['Age'].isnull().sum())

#Problem-3
embarked_encoded = pd.get_dummies(titanic_df['Embarked'], prefix='Embarked')
titanic_df = pd.concat([titanic_df, embarked_encoded], axis=1)
titanic_df.drop(columns=['Embarked'], inplace=True)
print("Modified DataFrame:")
print(titanic_df.head())

#Problem-4
mean_survival_rates = titanic_df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_rates)
gender_survival = titanic_df.groupby('Sex')['Survived'].mean()

fig, ax = plt.subplots(figsize=(8, 6))
gender_survival.plot(kind='bar', ax=ax, color=['lightblue', 'lightcoral'])
ax.set_title('Survival Rates by Gender', fontsize=14)
ax.set_xlabel('Gender', fontsize=12)
ax.set_ylabel('Survival Rate', fontsize=12)
ax.set_xticklabels(['Male', 'Female'], rotation=0, fontsize=10)
ax.set_ylim(0, 1)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

gender_survival_counts = titanic_df.groupby(['Sex', 'Survived']).size().unstack(fill_value=0)
gender_survival_counts.plot(kind='bar', stacked=False, color=['lightblue', 'lightcoral'], figsize=(10, 6))
plt.title('Survival Distribution by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(['Did Not Survive', 'Survived'], title='Survival', fontsize=10)
plt.show()
```

```python
#Problem-5
import pandas as pd
import matplotlib.pyplot as plt

titanic = pd.read_csv("/content/drive/MyDrive/Datasets/Titanic-Dataset.csv")
grouped_data = titanic.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
print("Mean Survival Rates by Gender and Port of Embarkation:")
print(grouped_data)

ports = ['C', 'Q', 'S']
male_survival = grouped_data.loc[ports, 'male']
female_survival = grouped_data.loc[ports, 'female']
x = range(len(ports))

plt.figure(figsize=(10, 6))
bar_width = 0.4
plt.bar(x, male_survival, width=bar_width, color='blue', label='Male', align='center')
plt.bar([p + bar_width for p in x], female_survival, width=bar_width, color='pink', label='Female', align='center')
plt.title("Mean Survival Rates by Gender and Port of Embarkation", fontsize=16)
plt.xlabel("Port of Embarkation", fontsize=12)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.xticks([p + bar_width / 2 for p in x], ports, fontsize=10)
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
#Problem-6
import pandas as pd
import matplotlib.pyplot as plt

titanic = pd.read_csv("/content/drive/MyDrive/Datasets/Titanic-Dataset.csv")
titanic['AgeGroup'] = pd.qcut(titanic['Age'], q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])
grouped_data = titanic.groupby(['Pclass', 'AgeGroup'])['Survived'].mean().unstack()
print("Mean Survival Rates by Age Group and Pclass:")
print(grouped_data)

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', stacked=False, color=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'yellow'], figsize=(10, 6))

plt.title("Survival Rates by Age Group and Passenger Class", fontsize=16)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
