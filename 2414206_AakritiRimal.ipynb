1. Data Exploration and Understanding:

```python
import pandas as pd

#Dataset Overview:
data = pd.read_csv("/content/drive/MyDrive/Assignment 1/WHR-2024-5CS037.csv")

print("First 10 rows of the dataset:")
print(data.head(10))

num_rows, num_columns = data.shape
print(f"\nNumber of Rows: {num_rows}, Number of Columns: {num_columns}")

print("\nColumns and their data types:")
print(data.dtypes)

#Basic Statistics:

#Calculating mean, median and standard deviation for the Score column
mean_score = data['score'].mean()
median_score = data['score'].median()
std_dev_score = data['score'].std()

print("\nStatistics for 'score' column:")
print(f"Mean: {mean_score:.2f}")
print(f"Median: {median_score:.2f}")
print(f"Standard Deviation: {std_dev_score:.2f}")

#Identifying the country wih highest and lowest happiness score
highest_score = data.loc[data['score'].idxmax()]
lowest_score = data.loc[data['score'].idxmin()]

print("\nCountry with the highest happiness score:")
print(highest_score[['Country name', 'score']])

print("\nCountry with the lowest happiness score:")
print(lowest_score[['Country name', 'score']])

#Missing Values
missing_values = data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values[missing_values > 0])

#Filtering and Sorting
filtered_data = data[data['score'] > 7.5]
print("\nCountries with Score greater than 7.5:")
print(filtered_data)

# Sort the filtered data by GDP per Capita in descending order
sorted_data = filtered_data.sort_values(by='Log GDP per capita', ascending=False).head(10)
print("\nTop 10 countries with Score > 7.5 sorted by GDP per Capita:")
print(sorted_data[['Country name', 'score', 'Log GDP per capita']])

#Adding New Columns
def happiness_category(score):
    if score < 4:
        return 'Low'
    elif 4 <= score <= 6:
        return 'Medium'
    else:
        return 'High'

data['Happiness Category'] = data['score'].apply(happiness_category)

print("\nDataset with 'Happiness Category' column added:")
print(data[['Country name', 'score', 'Happiness Category']].head(10))
```

Task - 3 - Outlier Detection:

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = '/content/drive/MyDrive/Assignment 1/south_asia_dataset.csv'
south_asian_df = pd.read_csv(dataset)

#Identify outliers for Score and GDP per Capita using the 1.5 × IQR rule
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers_score = detect_outliers(south_asian_df, 'score')
outliers_gdp = detect_outliers(south_asian_df, 'Log GDP per capita')

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

plt.figure(figsize=(10, 6))

plt.scatter(south_asian_df['Log GDP per capita'], south_asian_df['score'], color='blue', label='Normal')

plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='purple', label='Outliers')

plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Outlier Detection: GDP per Capita vs Happiness Score')
plt.legend()
plt.grid(True)
plt.show()

print("Outlier countries based on Score and GDP per Capita:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])
```

Task - 4 - Exploring Trends Across Metrics:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

dataset = '/content/drive/MyDrive/Assignment 1/south_asia_dataset.csv'
south_asian_df = pd.read_csv(dataset)

#Calculating Pearson correlation between selected metrics and Score
freedom_corr, _ = pearsonr(south_asian_df['Freedom to make life choices'], south_asian_df['score'])
generosity_corr, _ = pearsonr(south_asian_df['Generosity'], south_asian_df['score'])

print(f"Pearson correlation between 'Freedom to Make Life Choices' and Score: {freedom_corr:.2f}")
print(f"Pearson correlation between 'Generosity' and Score: {generosity_corr:.2f}")

plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=south_asian_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'green'})
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
plt.title('Freedom to Make Life Choices vs Happiness Score')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=south_asian_df, scatter_kws={'color': 'green'}, line_kws={'color': 'purple'})
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
plt.title('Generosity vs Happiness Score')
plt.grid(True)
plt.show()

if freedom_corr > generosity_corr:
    strongest_metric = 'Freedom to Make Life Choices'
    weakest_metric = 'Generosity'
else:
    strongest_metric = 'Generosity'
    weakest_metric = 'Freedom to Make Life Choices'

print(f"\nThe strongest correlation with the Happiness Score is: {strongest_metric}")
print(f"The weakest correlation with the Happiness Score is: {weakest_metric}")
```

Strongest Relationship: A higher Pearson correlation value indicates a stronger connection. If Freedom to Make Life Choices shows a high correlation, it implies that countries with more personal freedoms tend to have higher happiness scores.

Weakest Relationship: A lower Pearson correlation value, like with Generosity, suggests a weaker relationship. This may mean that Generosity has less of a direct impact on happiness scores in South Asian countries, or its effect is more subtle compared to other factors.

Task - 5 - Gap Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = '/content/drive/MyDrive/Assignment 1/south_asia_dataset.csv'
south_asian_df = pd.read_csv(dataset)

#Add a new column GDP-Score Gap
south_asian_df['GDP-Score Gap'] = south_asian_df['Log GDP per capita'] - south_asian_df['score']

ranked_by_gap_desc = south_asian_df.sort_values(by='GDP-Score Gap', ascending=False)
ranked_by_gap_asc = south_asian_df.sort_values(by='GDP-Score Gap', ascending=True)

#Highlighting the top 3 countries with the largest positive and negative gaps
top_positive_gap = ranked_by_gap_desc.head(3)
top_negative_gap = ranked_by_gap_asc.head(3)

plt.figure(figsize=(10, 6))
plt.barh(top_positive_gap['Country name'], top_positive_gap['GDP-Score Gap'], color='pink', label='Top Positive Gaps')
plt.barh(top_negative_gap['Country name'], top_negative_gap['GDP-Score Gap'], color='lightgreen', label='Top Negative Gaps')

plt.xlabel('GDP-Score Gap')
plt.ylabel('Country')
plt.title('Top 3 Countries with Largest Positive and Negative GDP-Score Gaps')
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```

Analysis:


Reasons for Large Gaps:


Negative Gap: Even with a strong economy, issues like inequality, weak support systems, or political problems can lower happiness.


Positive Gap: Strong community bonds, good social support, or cultural factors can make people happy even with a smaller economy.

Implications:


A positive gap shows that happiness doesn’t depend only on GDP. Strong social and health services can boost happiness even with less economic strength.


A negative gap highlights that economic growth alone isn’t enough. Problems like inequality or corruption can stop people from feeling the benefits of wealth
